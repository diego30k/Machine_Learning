```python
# Una funcion que nos permita hacer dummies
def creardummies(df, var):
    dummy = pd.get_dummies(df[var], prefix = var)
    df = df.drop(var, axis = 1)
    df = pd.concat([df, dummy], axis = 1)
    return df
```

##### Hallar valores missing y reemplazarlos

```python
Q1 = df[var_num].quantile(0.25)
Q3 = df[var_num].quantile(0.75)
IQR = Q3 - Q1
IQR
```

```python
# % de missing a imputar
(((df[var_num] < (Q1 - 3 * IQR)) | (df[var_num] > (Q3 + 3 * IQR))).sum()/df[var_num].shape[0])
```

```python
# Eliminacion de todos los missing, incluyendo los ya existentes en la data
df_out = df[var_num][~(((df[var_num] < (Q1 - 3 * IQR)) | (df[var_num] > (Q3 + 3 * IQR)))).any(axis=1)]
df_out.shape
```

```python
df[var_num].shape, df_out.shape
```

```python
ser_out = (((df[var_num] < (Q1 - 3 * IQR)) | (df[var_num] > (Q3 + 3 * IQR))).sum()/df[var_num].shape[0])
```

```python
ser_out
```

```python
var_out = list(ser_out[ser_out > 0].index)
var_out
```

```python
from time import time
time_star = time()

for v in var_out:
    df[v] = df[v].mask((df[v] < (float(df[v].quantile([.25])) - 3 * (float(df[v].quantile([.75])) - float(df[v].quantile([.25]))))) | 
                       (df[v] > (float(df[v].quantile([.75])) + 3 * (float(df[v].quantile([.75])) - float(df[v].quantile([.25]))))))

time_end = time()
print ("Time: ", np.round((time_end-time_star)/60,2), " minutes")
```
