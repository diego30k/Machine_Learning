```python
import pandas as pd
import nltk

#CAMBIAR NOMBRE TIENDA
tienda = 'homedepot'
tweets = pd.read_csv(tienda + ".csv",sep="~",dtype={'id': str, 'retweeted_id': str})
tweets.head(50)
```

```python
# Seleccionar solo tweets en ingles
tweets = tweets[tweets['lang'] == "en"]

# Limpiado de usaurios
tweets = tweets[~tweets.user.isin([tienda,'HomeDepot','HomeDepotCareer','hrdivision69'])]
tweets = tweets[~tweets.retweeted_user.isin([tienda,'HomeDepot','HomeDepotCareer','hrdivision69'])]

# Limpieza de campo source
tweets['source_group'] = 'Others'
tweets.set_value(tweets.source == 'Twitter for Android', 'source_group', 'Android')
tweets.set_value(tweets.source == 'Twitter for iPhone', 'source_group', 'Iphone')
tweets.set_value(tweets.source == 'Twitter for iPad', 'source_group', 'Ipad')
tweets.set_value(tweets.source == 'Twitter Web Client', 'source_group', 'Web')

# Columna de tienda
tweets['store'] = tienda
```

```python
#Limpieza de tweets
from nltk.tokenize import TweetTokenizer
from nltk.corpus import stopwords
import re, string
import nltk
stopwords=stopwords.words('english')
english_vocab = set(w.lower() for w in nltk.corpus.words.words())

# Creamos funcion de limpieza de tweets
def process_tweet_text(tweet):
    if tweet.startswith('@null'):
        return "[Tweet not available]"
    tweet = re.sub(r'\$\w*','',tweet)
    tweet = re.sub(r'https?:\/\/.*\/\w*','',tweet)
    tweet = re.sub(r'['+string.punctuation+']+', ' ',tweet)
    twtok = TweetTokenizer(strip_handles=True, reduce_len=True)
    tokens = twtok.tokenize(tweet)
    tokens = [i.lower() for i in tokens if i not in stopwords and len(i) > 2 and i in english_vocab]
    return " ".join(tokens)

tweets['text_clean'] = tweets['text'].apply(lambda x: process_tweet_text(x)) 
tweets.head()
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()

tweets['sent_neg'] = tweets['text_clean'].apply(lambda x: sid.polarity_scores(x)['neg'])
tweets['sent_pos'] = tweets['text_clean'].apply(lambda x: sid.polarity_scores(x)['pos'])
tweets['sent_neu'] = tweets['text_clean'].apply(lambda x: sid.polarity_scores(x)['neu'])

tweets['sentimiento'] = 0
tweets.set_value(tweets.sent_neg > tweets.sent_pos, 'sentimiento', -1)
tweets.set_value(tweets.sent_pos > tweets.sent_neg, 'sentimiento', 1)
```

```python
# Word Cloud
from nltk.stem.wordnet import WordNetLemmatizer 
from nltk import word_tokenize

cloud = pd.DataFrame({'id' : [], 'word' : [], 'count' : []})

for i, row in tweets.iterrows():
    words = word_tokenize(row['text_clean'].lower())
    for w in words:
        tmp = pd.DataFrame({"id": str(row['id']), "word": [w], 'count': [1]})
        cloud = cloud.append(tmp,ignore_index=True)

lem = WordNetLemmatizer()
cloud["word_root"] = cloud["word"].apply(lambda x: lem.lemmatize(x, "v"))
```

```python
# Hashtags
hashtags = pd.DataFrame({'id' : [], 'hashtag' : [], 'count' : []})

for i, row in tweets.iterrows():
    words = [tag[0:] for tag in row['text'].split(' ') if tag.startswith("#")]
    for w in words:
        tmp = pd.DataFrame({"id": str(row['id']), "hashtag": [w], 'count': [1]})
        hashtags = hashtags.append(tmp,ignore_index=True)
```

```python
tweets.to_csv(tienda + '_procesed.csv', index=False, sep='~',encoding = 'utf-8')
cloud.to_csv(tienda + '_wordcloud.csv', index=False, sep='~')
hashtags.to_csv(tienda + '_hashtags.csv', index=False, sep='~')
```

```python
# OPCIONAL PARA CLUSTERING DE TWEETS
from gensim import corpora, models
from nltk.corpus import stopwords
from nltk.stem.wordnet import WordNetLemmatizer
import string
stop = set(stopwords.words('english'))
exclude = set(string.punctuation)
lemma = WordNetLemmatizer()
def clean(doc):
    stop_free = " ".join([i for i in doc.lower().split() if i not in stop])
    punc_free = ''.join(ch for ch in stop_free if ch not in exclude)
    normalized = " ".join(lemma.lemmatize(word) for word in punc_free.split())
    return normalized
texts = [text for text in tweets['text_clean'] if len(text) > 2]
doc_clean = [clean(doc).split() for doc in texts]
dictionary = corpora.Dictionary(doc_clean)
doc_term_matrix = [dictionary.doc2bow(doc) for doc in doc_clean]
ldamodel = models.ldamodel.LdaModel(doc_term_matrix, num_topics=5, id2word = 
dictionary, passes=5)
for topic in ldamodel.show_topics(num_topics=6, formatted=False, num_words=6):
    print("Topic {}: Words: ".format(topic[0]))
    topicwords = [w for (w, val) in topic[1]]
    print(topicwords)
```
